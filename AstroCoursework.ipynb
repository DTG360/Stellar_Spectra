```python
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
mastarall = fits.open(r'C:\Physics\Astro\mastarall-v2_4_3-v1_0_2.fits')
mastarall.info()
```

```python
mastarall[1].data.names
```

```python
ksel = np.logical_and(mastarall[1].data["INPUT_TEFF"]
> 3700 , mastarall[1].data["INPUT_TEFF"] < 5200)
plt.hist(mastarall[1].data[ksel]["INPUT_TEFF"],bins=10)
plt.title("teff")
plt.show()
```

```python
kstars = mastarall[1].data[ksel]
```

```python
plt.hist(kstars["INPUT_LOGG"])
plt.title("log g")
plt.show()
```

```python
len(kstars) #number of stars
```

```python
print(kstars[36]["MANGAID"])
```

```python
goodspec = fits.open(r'C:\Physics\Astro\mastar-goodspec-v2_4_3-v1_0_2.fits')[1].data
```

```python
visits = goodspec["MANGAID"] == kstars[36]["MANGAID"]
wl = goodspec[visits]["WAVE"][0]
flux = goodspec[visits]["FLUX"][0]
```
